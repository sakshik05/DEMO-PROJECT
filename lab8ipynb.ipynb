```python
# --- Setup: Creating a sample text file ---
with open("sample_input.txt", "w") as f:
    f.write("Hello world!\nThis is a file handling test.\nPython is fun.")

# --- a) Main Program ---
try:
    # 1. Read the original file
    with open("sample_input.txt", "r") as source_file:
        content = source_file.read()

    # 2. Convert to uppercase
    upper_content = content.upper()

    # 3. Write to the new file
    with open("uppercase_output.txt", "w") as target_file:
        target_file.write(upper_content)

    print("✅ Conversion complete! Contents of 'uppercase_output.txt':")
    print("-" * 20)
    print(upper_content)
    print("-" * 20)

except FileNotFoundError:
    print("❌ Error: The source file was not found.")
```

```python
# --- Setup: Creating a source Python script with comments ---
source_code = """# This is a header comment
import math

# Calculating the area
radius = 5  # Inline comment (will keep the code part)
area = math.pi * radius**2
print(area) # Print the result
"""
with open("myscript.py", "w") as f:
    f.write(source_code)

# --- Main Application ---
source_name = input("Enter source file name (e.g., myscript.py): ")
dest_name = input("Enter destination file name (e.g., clean_script.py): ")

try:
    with open(source_name, "r") as src, open(dest_name, "w") as dest:
        print(f"\n--- Original Content ({source_name}) ---")
        for line in src:
            print(line.rstrip()) # Print original for comparison

            # Remove lines that are purely comments (start with #)
            # or lines that are empty/whitespace
            clean_line = line.split('#')[0].rstrip()

            if clean_line: # Only write if there is actual code left
                dest.write(clean_line + "\n")

    # Display the result
    print(f"\n--- Copied Content without Comments ({dest_name}) ---")
    with open(dest_name, "r") as final:
        print(final.read())

except FileNotFoundError:
    print("❌ Error: One of the files could not be opened. Check the names.")
```
