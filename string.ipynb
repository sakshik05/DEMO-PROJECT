```python
# 1. Access and print characters from string
print("\n1. Access characters")
s = "Python"
for ch in s:
    print(ch)

# 2. Print string and extract characters
print("\n2. Extract characters")
print(s)
print(s[0], s[1])

# 3. Print words with their length
print("\n3. Words with length")
s = "Python is easy"
for word in s.split():
    print(word, len(word))

# 4. Print EVEN length words
print("\n4. Even length words")
for word in s.split():
    if len(word) % 2 == 0:
        print(word)

# 5. Count vowels
print("\n5. Count vowels")
s = "education"
vowels = "aeiouAEIOU"
count = sum(1 for ch in s if ch in vowels)
print(count)

# 6. Passing string to function
print("\n6. String to function")
def greet(name):
    print("Hello", name)
greet("Sakshi")

# 7. Multiple copies using *
print("\n7. Multiple copies")
print("Hi " * 3)

# 8. Appending using +=
print("\n8. Append using +=")
s = "Hello"
s += " World"
print(s)

# 9. Concatenate using +
print("\n9. Concatenate")
a = "Good"
b = "Morning"
c = a + " " + b
print(c)

# 10. Check substring using in
print("\n10. Substring check")
print("Py" in "Python")

# 11. Hexadecimal values in string
print("\n11. Hex value")
hex_str = "\x50\x79\x74\x68\x6F\x6E"
print(hex_str)

# 12. Print double quotes
print("\n12. Double quotes")
name = "Sakshi"
print(f'"{name}"')

# 13. Ignore escape sequence
print("\n13. Raw string")
print(r"C:\newfolder\test")

# 14. Number of substrings
print("\n14. Total substrings")
s = "abc"
n = len(s)
print(n*(n+1)//2)

# 15. Reverse string (5 ways)
print("\n15. Reverse 5 ways")
s = "Python"
print(s[::-1])
print("".join(reversed(s)))
res=""
for ch in s:
    res = ch + res
print(res)
print("".join([s[i] for i in range(len(s)-1,-1,-1)]))
print(list(s)[::-1])

# 16. Reverse using stack & reversed
print("\n16. Reverse using stack")
stack = list(s)
rev=""
while stack:
    rev += stack.pop()
print(rev)
print("".join(reversed(s)))

# 17. Split into characters
print("\n17. Split into characters")
print(list("Hello"))

# 18. Slicing
print("\n18. Slicing")
s = "Programming"
print(s[0:6])
print(s[-3:])

# 19. Repeat M chars N times
print("\n19. Repeat M chars N times")
s="Hello"
print(s[:2]*3)

# 20. Swap characters
print("\n20. Swap characters")
s="abcd"
lst=list(s)
lst[0],lst[1]=lst[1],lst[0]
print("".join(lst))

# 21. Remove character at index
print("\n21. Remove index")
s="Python"
i=2
print(s[:i]+s[i+1:])

# 22. Add fixed message
print("\n22. Add fixed message")
msg="Welcome"
print(msg+" to Python")

# 23. Matched characters in two strings
print("\n23. Matched characters")
s1="hello"
s2="world"
print(set(s1)&set(s2))

# 24. Frequency of each character
print("\n24. Frequency")
from collections import Counter
print(Counter("banana"))

# 25. Extract mobile number
print("\n25. Extract mobile")
import re
text="Call me at 9876543210"
print(re.findall(r'\d{10}',text))

# 26. Replace special string
print("\n26. Replace string")
para="I like Java"
print(para.replace("Java","Python"))

# 27. ASCII values
print("\n27. ASCII values")
for ch in "ABC":
    print(ch, ord(ch))

# 28. Reverse string with digits
print("\n28. Reverse digits string")
s="12345"
print(s[::-1])

# 29. Convert to camelCase
print("\n29. camelCase")
s="hello world python"
words=s.split()
camel=words[0]+"".join(w.capitalize() for w in words[1:])
print(camel)

# 30. Capitalize first letter
print("\n30. Capitalize words")
print("hello world".title())

# 31. Check palindrome
print("\n31. Palindrome check")
s="madam"
print("Palindrome" if s==s[::-1] else "Not Palindrome")

# 32. Count uppercase and lowercase
print("\n32. Upper & Lower count")
s="PyTHon"
u=l=0
for ch in s:
    if ch.isupper(): u+=1
    elif ch.islower(): l+=1
print("Upper:",u,"Lower:",l)

# 33. Count letters and digits
print("\n33. Letters & Digits")
s="Python123"
letters=digits=0
for ch in s:
    if ch.isalpha(): letters+=1
    elif ch.isdigit(): digits+=1
print("Letters:",letters,"Digits:",digits)

# 34. List of chars to string
print("\n34. List to string")
lst=['P','y','t','h','o','n']
print("".join(lst))

# 35. Check variable is string
print("\n35. Check string type")
x="Hello"
print(isinstance(x,str))

# 36. Count occurrence of word
print("\n36. Word count")
text="python is easy python"
print(text.split().count("python"))

# 37. Search pattern
print("\n37. Search pattern")
print("easy" in text)

# 38. Remove i-th character
print("\n38. Remove i-th char")
s="Hello"
i=1
print(s[:i]+s[i+1:])

# 39. Length of string (different ways)
print("\n39. Length methods")
s="Python"
print(len(s))
count=0
for _ in s: count+=1
print(count)

# 40. String contains all vowels
print("\n40. Contains all vowels")
s="aeiouxyz"
print(all(v in s for v in "aeiou"))

# 41. Least frequent character
print("\n41. Least frequent char")
cnt=Counter("banana")
print(min(cnt,key=cnt.get))

# 42. Split and join
print("\n42. Split & Join")
s="Python is fun"
print("-".join(s.split()))

# 43. Words greater than length k
print("\n43. Words > k length")
s="Python is very powerful language"
k=6
for word in s.split():
    if len(word)>k:
        print(word)

# 44. Check special characters
print("\n44. Special characters")
import string
s="Hello@123"
print(any(ch in string.punctuation for ch in s))

# 45. Max frequency character
print("\n45. Max frequency char")
cnt=Counter("banana")
print(max(cnt,key=cnt.get))

# 46. Check binary string
print("\n46. Binary string check")
s="101010"
print(all(ch in "01" for ch in s))

# 47. Uncommon words from two strings
print("\n47. Uncommon words")
s1="apple banana mango"
s2="banana orange mango"
set1=set(s1.split())
set2=set(s2.split())
print(set1^set2)
```
