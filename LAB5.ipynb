```python
# --- Lab Assignment 1 ---

# Input: Asking user for a series of integers
user_input = input("Enter a series of integers separated by space: ")
# Convert input string into a tuple of integers
my_tuple = tuple(int(x) for x in user_input.split())

print("\n--- Results ---")

# a) Total number of items
print(f"a) Total number of items: {len(my_tuple)}")

# b) Last item in the tuple
print(f"b) Last item: {my_tuple[-1]}")

# c) Reverse order
print(f"c) Elements in reverse: {my_tuple[::-1]}")

# d) Check for integer 5
if 5 in my_tuple:
    print("d) Contains 5? Yes")
else:
    print("d) Contains 5? No")

# e) Remove first/last, sort, and print
# Note: Since tuples are immutable, we slice and convert to list to sort
if len(my_tuple) > 2:
    remaining_items = list(my_tuple[1:-1])
    remaining_items.sort()
    print(f"e) Sorted remaining items: {remaining_items}")
else:
    print("e) Not enough items to remove first and last.")
```

```python
# --- Lab Assignment 2 ---

# Input: Prices of sold items
price_input = input("Enter the prices of sold items separated by space: ")
prices = tuple(float(x) for x in price_input.split())

print("\n--- Sales Report ---")

# a) Total number of items sold
print(f"a) Total items sold: {len(prices)}")

# b) Cheapest item sold
print(f"b) Price of cheapest item: {min(prices)}")

# c) Costliest item sold
costliest = max(prices)
print(f"c) Price of costliest item: {costliest}")

# d) Price list in ascending order
# sorted() returns a new list
print(f"d) Prices in ascending order: {sorted(prices)}")

# e) Number of costliest items sold
count_costliest = prices.count(costliest)
print(f"e) Number of costliest items sold: {count_costliest}")
```
