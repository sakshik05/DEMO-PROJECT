```python
print("--- TASK-1: Patterns ---")

# 1. Row-based increment (1, 12, 123...)
print("\nPattern 1:")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end="")
    print()

# 2. Repeated digits (1, 22, 333...)
print("\nPattern 2:")
for i in range(1, 6):
    for j in range(i):
        print(i, end="")
    print()

# 3. Decreasing row numbers (1, 21, 321...)
print("\nPattern 3:")
for i in range(1, 6):
    for j in range(i, 0, -1):
        print(j, end="")
    print()

# 4. Binary/Modulo pattern (1, 10, 101...)
print("\nPattern 4:")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j % 2, end="")
    print()

# 5. Continuous even numbers
print("\nPattern 5 (Even Triangle):")
num = 2
for i in range(1, 5):
    for j in range(i):
        print(num, end=" ")
        num += 2
    print()

# 6. Star pattern
print("\nPattern 6:")
for i in range(1, 6):
    print("*" * i)
```

```python
print("--- TASK-2: String Patterns ---")

# 1. Alphabet Triangle (A, AB, ABC...)
print("\nPattern 1:")
for i in range(1, 6):
    for j in range(i):
        print(chr(65 + j), end="") # 65 is 'A'
    print()

# 2. Alternating Star and Hash
print("\nPattern 2:")
for i in range(1, 6):
    for j in range(i):
        if j % 2 == 0:
            print("*", end=" ")
        else:
            print("#", end=" ")
    print()

# 3. Repeated Word Characters (p, yy, ttt...)
print("\nPattern 3:")
word = "python"
for i in range(len(word)):
    print(word[i] * (i + 1))

# 4. "python" Slicing Pattern
print("\nPattern 4:")
for i in range(1, len(word) + 1):
    print(word[:i])
```

```python
print("--- TASK-3: Consecutive Integers ---")

# Read n from STDIN
n = int(input("Enter an integer n: "))

# Print 1 to n without using string methods
for i in range(1, n + 1):
    print(i, end="")
print() # Final newline for formatting
```

```python
# Input Format: An integer n
n = int(input("Enter n: "))

# Constraints: 1 <= n <= 150
if 1 <= n <= 150:
    for i in range(1, n + 1):
        # end="" prevents a newline or space after each number
        print(i, end="")
    print() # Prints a newline at the very end
else:
    print("Please enter a number between 1 and 150.")
```

```python
print("--- TASK-4 ---")

# Using a loop starting from 5 down to 1
for i in range(5, 0, -1):
    # Python allows multiplying strings: "*" * 5 = "*****"
    print("*" * i)
```

```python
print("--- TASK-5 ---")

n = 5 # Number of rows
for i in range(1, n + 1):
    # Calculate spaces: (Total Rows - Current Row)
    spaces = " " * (n - i)

    # Calculate stars: (2 * Current Row - 1)
    stars = "*" * (2 * i - 1)

    print(spaces + stars)
```

```python
# Sideways Triangle Star Pattern
n = 5  # Peak number of stars

print("--- Sideways Triangle Pattern ---")

# Part 1: Increasing stars (1 to 5)
for i in range(1, n + 1):
    print("* " * i)

# Part 2: Decreasing stars (4 to 1)
for i in range(n - 1, 0, -1):
    print("* " * i)
```

```python
# Prime Number Finder
print("--- TASK 7: Prime Number Finder ---")

# Step 1: Using a while loop for input validation
while True:
    try:
        lower = int(input("Enter lower range: "))
        upper = int(input("Enter upper range: "))

        if lower < upper:
            break
        else:
            print("❌ Invalid Range: Lower number must be less than upper number.")
    except ValueError:
        print("❌ Error: Please enter valid integers only.")

print(f"\nPrime numbers between {lower} and {upper} are:")

# Step 2: Use a for loop to iterate through the range
for num in range(lower, upper + 1):
    if num > 1:  # Primes must be greater than 1
        is_prime = True
        # Step 3: Use a for loop to check for primality
        for i in range(2, int(num**0.5) + 1):
            if (num % i) == 0:
                is_prime = False
                break

        if is_prime:
            print(num, end=" ")
```
