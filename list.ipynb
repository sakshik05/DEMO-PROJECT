```python
# 1. Largest number from a list
print("\n1. Largest number")
lst = [10, 45, 23, 89, 12]
print("Largest:", max(lst))

# 2. Remove even numbers
print("\n2. Remove even numbers")
lst = [1, 2, 3, 4, 5, 6]
result = [x for x in lst if x % 2 != 0]
print(result)

# 3. Reverse a list
print("\n3. Reverse list")
lst = [1, 2, 3, 4, 5]
print(lst[::-1])

# 4. Square each element
print("\n4. Square elements")
lst = [1, 2, 3, 4]
print([x**2 for x in lst])

# 5. Remove empty element
print("\n5. Remove empty element")
num = ["Hello", 34, 45, "", 40]
result = [x for x in num if x != ""]
print("Result:", result)

# 6. Items divisible by 5
print("\n6. Divisible by 5")
lst = [10, 12, 15, 17, 20]
print([x for x in lst if x % 5 == 0])

# 7. Remove repetitive items
print("\n7. Remove duplicates")
num = [2,3,4,5,2,6,3,2]
result = []
for x in num:
    if x not in result:
        result.append(x)
print("Result:", result)

# 8. Append second list to first
print("\n8. Append list2 to list1")
list1 = [23, 24, 25, 26]
list2 = [27, 28, 29, 30]
list1.extend(list2)
print(list1)

# 9. Filter odd and even
print("\n9. Odd and Even")
lst = [2, 23, 24, 51, 46, 67]
even = [x for x in lst if x % 2 == 0]
odd = [x for x in lst if x % 2 != 0]
print("Even:", even, "Odd:", odd)

# 10. Sort elements
print("\n10. Sorted list")
lst = [5, 2, 8, 1]
print(sorted(lst))

# 11. Count items in list
print("\n11. Count items")
lst = [1,2,3,4,5]
print("Count:", len(lst))

# 12. Reverse list without built-in (indexing)
print("\n12. Reverse without built-in")
lst = ['a','b','c','d']
rev = []
for i in range(len(lst)-1, -1, -1):
    rev.append(lst[i])
print(rev)

# 13. Add 7000 after 6000
print("\n13. Add 7000 after 6000")
list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]
list1[2][2].insert(2, 7000)
print(list1)

# 14. Replace first 20 with 200
print("\n14. Replace 20 with 200")
list1 = [5, 10, 15, 20, 25, 50, 20]
if 20 in list1:
    index = list1.index(20)
    list1[index] = 200
print(list1)

# 15. Find range (max - min)
print("\n15. Range of list")
lst = [10, 2, 30, 4]
print("Range:", max(lst) - min(lst))

# 16. Insert element at specific position
print("\n16. Insert element")
lst = [1,2,3,4]
lst.insert(2, 99)
print(lst)

# 17. Common elements between two lists
print("\n17. Common elements")
list1 = [1,2,3,4]
list2 = [3,4,5,6]
common = list(set(list1) & set(list2))
print(common)

# 18. Second smallest element
print("\n18. Second smallest")
lst = [5, 1, 4, 2, 3]
lst = sorted(set(lst))
print(lst[1])

# 19. Remove elements at even indices
print("\n19. Remove even indices")
lst = [10,20,30,40,50]
result = [lst[i] for i in range(len(lst)) if i % 2 != 0]
print(result)

# 20. Sum of elements at odd indices
print("\n20. Sum of odd indices")
lst = [10,20,30,40,50]
sum_odd = sum(lst[i] for i in range(len(lst)) if i % 2 != 0)
print(sum_odd)
```
