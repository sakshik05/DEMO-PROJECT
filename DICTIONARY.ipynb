```python
# 1. Add a key to a dictionary
print("\n1. Add a key to dictionary")
d = {0: 10, 1: 20}
d[2] = 30
print(d)

# 2. Concatenate dictionaries
print("\n2. Concatenate dictionaries")
dic1 = {1:10, 2:20}
dic2 = {3:30, 4:40}
dic3 = {5:50, 6:60}
result = {}
for dic in (dic1, dic2, dic3):
    result.update(dic)
print(result)

# 3. Check whether a key exists
print("\n3. Check if key exists")
d = {1:10, 2:20}
key = 2
print("Key exists" if key in d else "Key not exists")

# 4. Iterate over dictionary
print("\n4. Iterate over dictionary")
d = {"a":1, "b":2}
for key, value in d.items():
    print(key, value)

# 5. Generate dictionary (x, x*x) for n
print("\n5. Dictionary (x, x*x)")
n = 5
d = {x: x*x for x in range(1, n+1)}
print(d)

# 6. Dictionary 1 to 15 with squares
print("\n6. Squares from 1 to 15")
d = {x: x*x for x in range(1,16)}
print(d)

# 7. Merge two dictionaries
print("\n7. Merge two dictionaries")
d1 = {"a":100}
d2 = {"b":200}
d1.update(d2)
print(d1)

# 8. Iterate using for loop
print("\n8. Iterate dictionary")
d = {"x":10, "y":20}
for k in d:
    print(k, d[k])

# 9. Sum of dictionary values
print("\n9. Sum of values")
d = {"a":10,"b":20,"c":30}
print(sum(d.values()))

# 10. Multiply dictionary values
print("\n10. Multiply values")
d = {"a":2,"b":3,"c":4}
mul = 1
for v in d.values():
    mul *= v
print(mul)

# 11. Remove a key
print("\n11. Remove a key")
d = {"a":1,"b":2}
d.pop("a")
print(d)

# 12. Map two lists into dictionary
print("\n12. Map lists into dictionary")
keys = ["a","b","c"]
values = [1,2,3]
print(dict(zip(keys, values)))

# 13. Sort dictionary by key
print("\n13. Sort by key")
d = {3:30,1:10,2:20}
print(dict(sorted(d.items())))

# 14. Max and Min values
print("\n14. Max and Min values")
d = {"a":5,"b":10,"c":2}
print("Max:", max(d.values()))
print("Min:", min(d.values()))

# 15. Dictionary from object fields
print("\n15. Dictionary from object")
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
s = Student("Sakshi", 20)
print(vars(s))

# 16. Remove duplicates from dictionary values
print("\n16. Remove duplicate values")
d = {"a":1,"b":2,"c":1}
unique = {}
for k,v in d.items():
    if v not in unique.values():
        unique[k] = v
print(unique)

# 17. Check if dictionary empty
print("\n17. Check if empty")
d = {}
print("Empty" if not d else "Not Empty")

# 18. Combine dictionaries by adding common keys
print("\n18. Combine dictionaries by adding values")
from collections import Counter
d1 = {'a':100,'b':200,'c':300}
d2 = {'a':300,'b':200,'d':400}
print(Counter(d1) + Counter(d2))

# 19. Print distinct values
print("\n19. Distinct values")
data = [{"V":"S001"},{"V":"S002"},{"VI":"S001"},{"VI":"S005"},
        {"VII":"S005"},{"V":"S009"},{"VIII":"S007"}]
values = {list(d.values())[0] for d in data}
print("Unique Values:", values)

# 20. Combinations of letters
print("\n20. Combinations of letters")
d = {'1':['a','b'], '2':['c','d']}
for i in d['1']:
    for j in d['2']:
        print(i+j)

# 21. Highest 3 values
print("\n21. Highest 3 values")
d = {"a":10,"b":50,"c":30,"d":40}
top3 = sorted(d.values(), reverse=True)[:3]
print(top3)

# 22. Combine values in list of dictionaries
print("\n22. Combine values in list of dictionaries")
data = [{'item':'item1','amount':400},
        {'item':'item2','amount':300},
        {'item':'item1','amount':750}]
result = Counter()
for d in data:
    result[d['item']] += d['amount']
print(result)

# 23. Create dictionary from string (count letters)
print("\n23. Dictionary from string")
s = "SBJAIN"
count = {}
for ch in s:
    count[ch] = count.get(ch,0) + 1
print(count)

# 24. Print dictionary in table format
print("\n24. Dictionary in table format")
d = {"Name":"Sakshi","Age":20}
print("Key\tValue")
for k,v in d.items():
    print(k,"\t",v)

# 25. Count values associated with key
print("\n25. Count values associated with key")
d = {"a":[1,2,3,4,5,6], "b":[1,2]}
print(len(d["a"]))
print(len(d["b"]))

# 26. Convert list into nested dictionary
print("\n26. List into nested dictionary")
lst = [1,2,3,4]
d = current = {}
for num in lst:
    current[num] = {}
    current = current[num]
print(d)

# 27. Sort list inside dictionary
print("\n27. Sort list in dictionary")
d = {"letters":["b","d","a","c"]}
d["letters"].sort()
print(d)

# 28. Remove spaces from dictionary keys
print("\n28. Remove spaces from keys")
d = {"first name":"Sakshi","last name":"Kalbande"}
new_d = {k.replace(" ",""):v for k,v in d.items()}
print(new_d)

# 29. Top three items in shop
print("\n29. Top 3 items")
shop = {'item1':45.50,'item2':35,'item3':41.30,'item4':55,'item5':24}
top3 = sorted(shop.items(), key=lambda x:x[1], reverse=True)[:3]
for item, price in top3:
    print(item, price)
```
