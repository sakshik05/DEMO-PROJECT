```python
import numpy as np
import statistics

# ---------------- MATRIX OPERATIONS ----------------
def matrix_operations():
    try:
        r = int(input("Enter number of rows: "))
        c = int(input("Enter number of columns: "))

        print("Enter elements for Matrix A:")
        A = []
        for i in range(r):
            row = list(map(float, input().split()))
            A.append(row)

        print("Enter elements for Matrix B:")
        B = []
        for i in range(r):
            row = list(map(float, input().split()))
            B.append(row)

        A = np.array(A)
        B = np.array(B)

        print("\n1. Addition\n2. Subtraction\n3. Multiplication")
        ch = int(input("Choose operation: "))

        if ch == 1:
            print("Result:\n", A + B)
        elif ch == 2:
            print("Result:\n", A - B)
        elif ch == 3:
            print("Result:\n", np.dot(A, B))
        else:
            print("Invalid choice")

    except Exception as e:
        print("Invalid Input:", e)


# ---------------- STATISTICS ----------------
def statistics_operations():
    try:
        data = list(map(float, input("Enter numbers separated by space: ").split()))

        print("Mean:", statistics.mean(data))
        print("Median:", statistics.median(data))
        print("Standard Deviation:", statistics.stdev(data))

    except statistics.StatisticsError:
        print("Not enough data to calculate statistics.")
    except Exception:
        print("Invalid Input")


# ---------------- UNIT CONVERSION ----------------
def unit_conversion():
    try:
        print("\n1. Meter to Kilometer")
        print("2. Kilometer to Meter")
        print("3. Celsius to Fahrenheit")
        print("4. Fahrenheit to Celsius")

        ch = int(input("Choose conversion: "))
        val = float(input("Enter value: "))

        if ch == 1:
            print("Result:", val / 1000, "km")
        elif ch == 2:
            print("Result:", val * 1000, "m")
        elif ch == 3:
            print("Result:", (val * 9/5) + 32, "F")
        elif ch == 4:
            print("Result:", (val - 32) * 5/9, "C")
        else:
            print("Invalid choice")

    except ZeroDivisionError:
        print("Division by Zero Error")
    except Exception:
        print("Invalid Input")


# ---------------- MAIN MENU ----------------
def main():
    while True:
        print("\n===== ENGI-CALC =====")
        print("1. Matrix Algebra")
        print("2. Statistical Analysis")
        print("3. Unit Conversion")
        print("4. Exit")

        try:
            choice = int(input("Enter your choice: "))

            if choice == 1:
                matrix_operations()
            elif choice == 2:
                statistics_operations()
            elif choice == 3:
                unit_conversion()
            elif choice == 4:
                print("Exiting Engi-Calc...")
                break
            else:
                print("Invalid Choice")

        except Exception:
            print("Please enter a valid number")


# Run program
main()
```

```python
import tkinter as tk
from tkinter import messagebox
import math

class EngineeringCalc:
    def __init__(self, root):
        self.root = root
        self.root.title("Engineering Calculator")
        self.equation = ""
        self.display_var = tk.StringVar()

        # Display Screen
        display = tk.Entry(root, textvariable=self.display_var, font=("Arial", 20), bd=10, relief="flat", justify='right')
        display.grid(row=0, column=0, columnspan=5, padx=10, pady=20)

        # Button definitions (Text, Row, Column)
        buttons = [
            ('sin', 1, 0), ('cos', 1, 1), ('tan', 1, 2), ('log', 1, 3), ('ln', 1, 4),
            ('(', 2, 0), (')', 2, 1), ('^', 2, 2), ('sqrt', 2, 3), ('clear', 2, 4),
            ('7', 3, 0), ('8', 3, 1), ('9', 3, 2), ('/', 3, 3), ('del', 3, 4),
            ('4', 4, 0), ('5', 4, 1), ('6', 4, 2), ('*', 4, 3), ('pi', 4, 4),
            ('1', 5, 0), ('2', 5, 1), ('3', 5, 2), ('-', 5, 3), ('e', 5, 4),
            ('0', 6, 0), ('.', 6, 1), ('exp', 6, 2), ('+', 6, 3), ('=', 6, 4)
        ]

        for (text, r, c) in buttons:
            cmd = lambda x=text: self.on_click(x)
            tk.Button(root, text=text, width=6, height=2, command=cmd).grid(row=r, column=c, padx=2, pady=2)

    def on_click(self, key):
        if key == '=':
            try:
                # Replace visual symbols with python-compatible math calls
                expr = self.equation.replace('^', '**').replace('sqrt', 'math.sqrt')
                expr = expr.replace('sin', 'math.sin(math.radians').replace('cos', 'math.cos(math.radians')
                expr = expr.replace('tan', 'math.tan(math.radians').replace('log', 'math.log10')
                expr = expr.replace('ln', 'math.log').replace('pi', 'math.pi').replace('e', 'math.e')

                # Close radians parenthesis for trig if needed
                if 'radians' in expr: expr += ')'

                result = eval(expr)
                self.display_var.set(result)
                self.equation = str(result)
            except:
                messagebox.showerror("Error", "Invalid Input")
                self.equation = ""
        elif key == 'clear':
            self.equation = ""
        elif key == 'del':
            self.equation = self.equation[:-1]
        else:
            self.equation += str(key)

        self.display_var.set(self.equation)

if __name__ == "__main__":
    root = tk.Tk()
    EngineeringCalc(root)
    root.mainloop()
```
