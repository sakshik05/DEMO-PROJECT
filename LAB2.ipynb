```python
# Program 1: While Loop
count = 1
while count <= 20:
    print(f"{count}. Hello Students")
    count += 1
```

```python
# Program 2: Sum of n numbers
n = int(input("Enter the limit: "))
total_sum = 0

for i in range(1, n + 1):
    total_sum += i

print(f"The sum of numbers from 1 to {n} is: {total_sum}")
```

```python
# Program 3: Armstrong Number
num = int(input("Enter a number: "))
order = len(str(num)) # Number of digits
temp = num
sum_pow = 0

while temp > 0:
    digit = temp % 10
    sum_pow += digit ** order
    temp //= 10

if num == sum_pow:
    print(f"{num} is an Armstrong number.")
else:
    print(f"{num} is NOT an Armstrong number.")
```

```python
# Program 4: Multiplication Table
num = int(input("Enter a number for the table: "))

print(f"\nMultiplication Table for {num}:")
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

```python
print("--- Pattern 1 (Floyd's Variant) ---")
k = 1
for i in range(1, 5):
    for j in range(i):
        print(k, end=" ")
        k += 1
    print()

print("\n--- Pattern 2 (Stars) ---")
for i in range(1, 6):
    print("*" * i)

print("\n--- Pattern 3 (Inverted Repeat) ---")
for i in range(1, 6):
    print((str(i) + " ") * (6 - i))

print("\n--- Pattern 4 (Reverse Range) ---")
for i in range(5, 0, -1):
    for j in range(i, 0, -1):
        print(j, end=" ")
    print()

print("\n--- Pattern 5 (Right-Aligned Numbers) ---")
for i in range(1, 6):
    # Spaces for alignment
    print("  " * (5 - i), end="")
    for j in range(1, i + 1):
        print(j, end=" ")
    print()

print("\n--- Pattern 6 (Star Pyramid) ---")
n = 5
for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)
```
