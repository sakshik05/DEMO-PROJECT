```python
# --- Lab Assignment 1 ---
def string_statistics():
    user_string = input("Enter a string: ")

    vowels = "aeiouAEIOU"
    v_count = 0
    c_count = 0
    s_count = 0
    l_count = 0

    for char in user_string:
        # a) Count Vowels
        if char in vowels:
            v_count += 1

        # b) Count Consonants (Must be a letter and NOT a vowel)
        if char.isalpha() and char not in vowels:
            c_count += 1

        # c) Count Spaces
        if char.isspace():
            s_count += 1

        # d) Count Lowercase Letters
        if char.islower():
            l_count += 1

    print("\n--- String Statistics ---")
    print(f"a) Number of Vowels:            {v_count}")
    print(f"b) Number of Consonants:        {c_count}")
    print(f"c) Number of Spaces:            {s_count}")
    print(f"d) Number of Lowercase Letters: {l_count}")

# Run the function
string_statistics()
```

```python
# --- Lab Assignment 2 ---
def capitalize_sequence():
    print("Enter your lines of text.")
    print("When you are finished, press Enter on an empty line:")

    lines = []

    while True:
        line = input()
        if line:
            # Add the capitalized version to our list
            lines.append(line.upper())
        else:
            # Exit loop if the user enters a blank line
            break

    print("\n--- Output ---")
    for capitalized_line in lines:
        print(capitalized_line)

# Run the function
capitalize_sequence()
```
